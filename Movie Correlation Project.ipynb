```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) # Adjust the configuration of the plots will create

df = pd.read_csv(r'D:\movies.csv')
```

```python
df.head()
```

```python
for col in df.columns: 
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df.dtypes
```

```python
df['budget'] = df['budget'].fillna(0).astype('int64')

df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
df.head()
```

```python

df['countrycorrect'] = df['released'].str.extract(r'\((.*?)\)$')
```

```python
df.head()
```

```python
df['yearcorrect'] = df['released'].str.extract(r'.*, (\d{4})')
```

```python
df.head()
```

```python
df = df.sort_values('gross', inplace=False, ascending=False)
```

```python
pd.set_option('display.max.rows', None)
```

```python
df.head().drop_duplicates().sort_values(by='gross', ascending=False)
```

```python
df.head()
```

```python
# Scatter plot with Budget vs Gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget For Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
numeric_columns = df.select_dtypes(include=['number'])

# Compute the correlation matrix
correlation_matrix = numeric_columns.corr('pearson')
print(correlation_matrix)
```

```python

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

# Display the resulting DataFram
df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movei Features')
plt.show()
```
